```python
from pynsn import *
from pynsn.image import pil_image
```

To generate a random nsn stimulus, you have to define

1. an NSN factory
2. a distribution for the object size 

```python
# Generate random nsn stimulus with 15 dots
stim = NSNStimulus(target_area_shape=Dot(300))
stim.add_shapes(Dot(diameter=20, attribute="darkgreen"), n=12, random_position=True)
```

Make picture of the stimulus

```python
pil_image.create(stim)
```

## Adapting Stimulus Colour

```python
stim.colours.target_area="#FFF0D9"
stim.colours.convex_hull="red"
stim.colours.center_of_mass="lightblue"
stim.colours.center_of_field_area="lightgreen"
stim.colours.background="#3D3D3D"

pil_image.create(stim)
```

## Incremental 

```python
stim = NSNStimulus(target_area_shape=Dot(100, attribute="darkgray"))
for n in range(10):
    stim.add_shapes(Rectangle(size=[9, 12], attribute="darkmagenta"),random_position=True)
    img = pil_image.create(stim)
    display(img)
    
```
